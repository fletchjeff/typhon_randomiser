```python
inport = mido.open_input('Typhon')
outport = mido.open_output('Typhon')
```

```python
def padded_hex(number_to_pad,pad_size):
    hex_string = str(hex(number_to_pad)).split("x")[1].upper()
    hex_string_len = len(hex_string)
    if hex_string_len >= pad_size:
        return hex_string
    else:
        for i in range(pad_size - hex_string_len):
            hex_string = "0" + hex_string
        return hex_string
```

```python
get_patch_1 = mido.Message('sysex', data=[0, 33, 53, 0, 4, 8, 1, 0, 0, 0])
outport.send(get_patch_1)
patch_1 = inport.receive()
```

```python
get_patch_2 = mido.Message('sysex', data=[0, 33, 53, 0, 4, 8, 1, 0, 0, 0])
outport.send(get_patch_2)
patch_2 = inport.receive()
```

```python
sys_1_hex = patch_1.hex().split(" ")
sys_2_hex = patch_2.hex().split(" ")
file_size = round(max(len(patch_2.data),len(patch_1.data))/16)
for i in range(file_size):
    hexout = padded_hex(i,4) + ": "
    for j in range(16):
        if ((i*16) + j) < len(sys_1_hex):
            hexout = hexout + sys_1_hex[(i*16) + j] + " " 
        else:
            hexout = hexout + "**" + " " 
    hexout = hexout + " --  "
    for j in range(16):
        if ((i*16) + j) < len(sys_2_hex):
            if sys_1_hex[(i*16) + j] == sys_2_hex[(i*16) + j]:
                hexout = hexout + sys_2_hex[(i*16) + j] + " " 
            else:
                hexout = hexout + sys_2_hex[(i*16) + j] + "*"
        else:
            hexout = hexout + "**" + " "             
    print(hexout)
```
